```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.tsa.holtwinters import SimpleExpSmoothing

# Load the CSV file
file_path = r"C:\Users\CHARUSAT\Documents\Business_forecasting\New folder (6)\vgsales.csv"
df = pd.read_csv(file_path)

# Apply Simple Exponential Smoothing
model = SimpleExpSmoothing(df['Global_Sales']).fit(smoothing_level=0.5, optimized=False)
df['Smoothed_Sales'] = model.fittedvalues

# Plot using Seaborn
sns.lineplot(x='Year', y='Global_Sales', data=df, label='Actual Sales', marker='o')
sns.lineplot(x='Year', y='Smoothed_Sales', data=df, label='Smoothed Sales', marker='x')

# Add labels and title
plt.title('Global Sales Trend with Simple Exponential Smoothing')
plt.xlabel('Year')
plt.ylabel('Global Sales (in millions)')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```
